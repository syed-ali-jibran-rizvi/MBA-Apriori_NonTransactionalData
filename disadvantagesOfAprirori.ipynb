```python
# Step 1: One-hot encode categorical columns and binary columns as well
df_onehot = pd.get_dummies(df[['main_category', 'sub_category', 'rating_category']], prefix=['main_category', 'sub_category', 'rating_category'])

# Step 2: Ensure all numerical features are binary (no_of_ratings, discount_price, actual_price)
df_onehot['no_of_ratings'] = df['no_of_ratings'].apply(lambda x: 1 if x > 0 else 0)
df_onehot['discount_price'] = df['discount_price'].apply(lambda x: 1 if x >= mean_discount_price else 0)
df_onehot['actual_price'] = df['actual_price'].apply(lambda x: 1 if x >= mean_actual_price else 0)

# Combine the new one-hot encoded columns into the transactional dataset
transactional_df = df_onehot
```

```python
transactional_df
```

```python
from mlxtend.frequent_patterns import apriori

# Apply apriori with a minimum support threshold of 0.3
frequent_itemsets = apriori(transactional_df, min_support=0.3, use_colnames=True)

# Display the frequent itemsets
print("Frequent Itemsets:")
print(frequent_itemsets)
```

```python
from mlxtend.frequent_patterns import association_rules

# Generate association rules with a minimum confidence of 0.7
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7, num_itemsets=len(frequent_itemsets))

# Display the association rules
print("\nAssociation Rules:")
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
# Display the association rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```

```python
# Format antecedents and consequents into string representation
rules['rule'] = rules['antecedents'].apply(lambda x: ', '.join(list(x))) + ' -> ' + rules['consequents'].apply(lambda x: ', '.join(list(x)))

# Display the formatted rules
print(rules[['rule', 'support', 'confidence', 'lift']])
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Initialize an empty graph
G = nx.DiGraph()

# Add nodes for each item in antecedents and consequents
for index, row in rules.iterrows():
    antecedents = ', '.join(list(row['antecedents']))
    consequents = ', '.join(list(row['consequents']))

    # Add nodes for antecedents and consequents
    G.add_node(antecedents)
    G.add_node(consequents)

    # Add edges with confidence as weight
    G.add_edge(antecedents, consequents, weight=row['confidence'])

# Draw the network
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G)  # Positions for nodes
nx.draw_networkx_nodes(G, pos, node_size=2000, node_color='skyblue')
nx.draw_networkx_edges(G, pos, width=2, alpha=0.5, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): f"Conf: {d['weight']:.2f}" for u, v, d in G.edges(data=True)})

plt.title('Association Rules Network')
plt.axis('off')  # Hide axis
plt.show()
```

```python
# Sort the rules by lift in descending order
rules_sorted = rules.sort_values('lift', ascending=False)

# Plot the top 10 rules by lift
plt.figure(figsize=(12, 8))
plt.barh(rules_sorted.head(10)['rule'], rules_sorted.head(10)['lift'], color='lightcoral')
plt.title('Top 10 Association Rules by Lift')
plt.xlabel('Lift')
plt.ylabel('Rules')
plt.show()
```
